# Mean and SD of fg_bg images

```
import glob
import cv2
import numpy as np
import os
from google.colab.patches import cv2_imshow
count = 0
sum_mean = [0, 0, 0]; sum_std = [0, 0, 0]
for i in glob.glob('Output/OverlayedImages/Gunjan/*/*.jpg'):
  img = cv2.imread(i)
  b = img[:,:,0]/255
  g = img[:,:,1]/255
  r = img[:,:,2]/255
  sum_mean[0] += np.mean(b)
  sum_std[0] += np.std(b)
  sum_mean[1] += np.mean(g)
  sum_std[1] += np.std(g)
  sum_mean[2] += np.mean(r)
  sum_std[2] += np.std(r)
  count=count+1
  if(count%10000==0):
    break
sum_mean[0] = sum_mean[0]/(count)
sum_std[0] = sum_std[0]/count
sum_mean[1] = sum_mean[1]/count
sum_std[1] = sum_std[1]/count
sum_mean[2] = sum_mean[2]/count
sum_std[2] = sum_std[2]/count
print("Mean: -",sum_mean)
print("stdDev: -", sum_std)
```

#Background Images

```
plt.figure(figsize=(40,40))
gs = gridspec.GridSpec(20,20)
gs.update(wspace=0.025, hspace=0.05)
count1 = 181;
for i in range(1,9):
  plt.subplot(gs[count1])
  img  = cv2.imread('Background/Shelves_'+str(i)+'.png')[:,:,::-1]
  pixels = np.array(img)
  plt.imshow(pixels)
  plt.axis('off')
  count1 = count1 + 1

plt.show()
```
