# Imports

```
import pickle
import matplotlib.pyplot as plt
```

# Set pickle file paths of all models

```
without_L1_and_L2 = '/content/gdrive/My Drive/Colab Notebooks/S6/Without_L1_and_L2/without_L1_and_L2.pickle'
with_L1 = '/content/gdrive/My Drive/Colab Notebooks/S6/With_L1/with_L1.pickle'
with_L2 = '/content/gdrive/My Drive/Colab Notebooks/S6/With_L2/with_L2.pickle'
with_L1_and_L2 = '/content/gdrive/My Drive/Colab Notebooks/S6/With_L1_and_L2/with_L1_and_L2.pickle'
```

# Load pickle object from each pickle file

```
with open(without_L1_and_L2, 'rb') as f1:
  without_L1_and_L2_obj = pickle.load(f1)

with open(with_L1, 'rb') as f2:
  with_L1_obj = pickle.load(f2)

with open(with_L2, 'rb') as f3:
  with_L2_obj = pickle.load(f3)

with open(with_L1_and_L2, 'rb') as f4:
  with_L1_and_L2_obj = pickle.load(f4)
```

# Plot validation accuracy change graph

```
val_acc_fig, axs = plt.subplots(1,1,figsize=(10,7))

axs.plot(without_L1_and_L2_obj['test_accuracy'], label='Without L1, L2')
axs.plot(with_L1_obj['test_accuracy'], label='With L1')
axs.plot(with_L2_obj['test_accuracy'], label='With L2')
axs.plot(with_L1_and_L2_obj['test_accuracy'], label='With L1, L2')
axs.set_title("Validation Accuracy Change")
axs.set_xlabel("Epoch")
axs.set_ylabel("Accuracy")
axs.legend(loc='lower right', prop={'size': 18})

val_acc_fig.savefig('/content/gdrive/My Drive/Colab Notebooks/S6/overall_validation_accuracy_change_plot')
```

# Plot validation loss change graph

```
val_loss_fig, axs = plt.subplots(1,1,figsize=(10,7))

axs.plot(without_L1_and_L2_obj['test_loss'], label='Without L1, L2')
axs.plot(with_L1_obj['test_loss'], label='With L1')
axs.plot(with_L2_obj['test_loss'], label='With L2')
axs.plot(with_L1_and_L2_obj['test_loss'], label='With L1, L2')
axs.set_title("Validation Loss Change")
axs.set_xlabel("Epoch")
axs.set_ylabel("Loss")
axs.legend(loc='upper right', prop={'size': 18})

val_loss_fig.savefig('/content/gdrive/My Drive/Colab Notebooks/S6/overall_validation_loss_change_plot')
```
